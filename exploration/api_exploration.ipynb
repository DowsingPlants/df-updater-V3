```python
import pandas as pd
import requests
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

# -----------------------------
# Pandas display setup
# -----------------------------
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 5)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)

# -----------------------------
# Headers (NBA requires browser-like headers)
# -----------------------------
DEFAULT_HEADERS = {
    "Host": "stats.nba.com",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/140.0.0.0 Safari/537.36",
    "Accept": "application/json, text/plain, */*",
    "Accept-Language": "en-US,en;q=0.5",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
    "Referer": "https://stats.nba.com/",
    "Pragma": "no-cache",
    "Cache-Control": "no-cache",
    "Sec-Ch-Ua": '"Chromium";v="140", "Google Chrome";v="140", "Not;A=Brand";v="24"',
    "Sec-Ch-Ua-Mobile": "?0",
    "Sec-Fetch-Dest": "empty",
    "Origin": "https://stats.nba.com",
    "Sec-Fetch-Site": "same-origin",
}

# -----------------------------
# Create session with retry logic
# -----------------------------
session = requests.Session()

retry_strategy = Retry(
    total=5,
    backoff_factor=1,
    status_forcelist=[429, 500, 502, 503, 504],
    allowed_methods=["GET"],
)

adapter = HTTPAdapter(max_retries=retry_strategy)
session.mount("https://", adapter)
session.headers.update(DEFAULT_HEADERS)

# -----------------------------
# Endpoint + parameters
# -----------------------------
url = "https://stats.nba.com/stats/boxscoreadvancedv3"

params = {
    "EndPeriod": 1,
    "EndRange": 0,
    "GameID": "0021700807",
    "RangeType": 0,
    "StartPeriod": 1,
    "StartRange": 0,
}

# -----------------------------
# Call API
# -----------------------------
response = session.get(url, params=params, timeout=30)
response.raise_for_status()

print("Status:", response.status_code)

data = response.json()

# -----------------------------
# Extract + Flatten Player Data
# -----------------------------
game_data = data["boxScoreAdvanced"]
game_id = game_data["gameId"]

all_rows = []

for team_side in ["homeTeam", "awayTeam"]:
    team = game_data[team_side]
    team_id = team["teamId"]
    team_tricode = team["teamTricode"]
    
    for player in team["players"]:
        base_info = {k: v for k, v in player.items() if k != "statistics"}
        stats = player["statistics"]
        
        flat_row = {
            **base_info,
            **stats,
            "gameId": game_id,
            "teamId": team_id,
            "teamTricode": team_tricode,
            "teamSide": team_side.replace("Team", "")
        }
        
        all_rows.append(flat_row)

df_players = pd.DataFrame(all_rows)

df_players.head()
```

```python
import sqlite3
import pandas as pd

PLAYER_ID = 203999


query_trad = """
SELECT *
FROM player_game_traditional
WHERE person_id = ?
"""

query_adv = """
SELECT *
FROM player_game_advanced
WHERE person_id = ?
"""

trad = pd.read_sql(query_trad, conn, params=[PLAYER_ID])
adv = pd.read_sql(query_adv, conn, params=[PLAYER_ID])

trad.head()
```

```python
df = trad.merge(
    adv,
    on=["game_id", "person_id"],
    suffixes=("", "_adv")
)

df.head()
```
